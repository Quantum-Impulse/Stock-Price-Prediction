```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")

sp500 = sp500.history(period="max")

sp500
```

```python
sp500.index
```

```python
sp500.plot.line(y="Close", use_index=True)
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500     # Tomorrow's Close is today's Close shifted by 1
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)   # Target is True if tomorrow's Close is greater than today's Close

sp500
```
